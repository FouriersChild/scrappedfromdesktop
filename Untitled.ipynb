```python
import gym
env = gym.make('Pendulum-v0')
env.reset()
for t in range(1000):
    env.render()
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
env.close()

# Set up environment
# Library involves the idea of selecting an action and recieving information back from environment
# ------------------------------
# 1. Render env
# 2. Select action
# 3. get info back
# 4. Repeat
```
