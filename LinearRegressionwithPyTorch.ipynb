```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
x = 10*torch.randn(100,1)
print(x)
y = x + 3*torch.randn(100,1)
print(y)
plt.plot(x.numpy(),y.numpy(), 'o')
```

```python
class LR(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.layer1 = nn.Linear(input_dim, output_dim)
    def forward(self, x):
        pred = self.layer1(x)
        return pred
```

```python
torch.manual_seed(1)
model = LR(1, 1)
```

```python
[w, b] = model.parameters()
def get_params():
    return (w[0][0].item(), b[0].item())
```

```python
import numpy as np
def plot_fit(title):
    plt.title = title
    w1, b1 = get_params()
    x1 = np.array([-30, 30])
    y1 = w1*x1 + b1
    plt.plot(x1, y1, 'r')
    plt.scatter(x, y)
    plt.show()
```

```python
plot_fit("Initial_Model")
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
```

```python
epochs = 100
losses = []
for i in range(epochs):
    y_pred = model.forward(x)
    loss = criterion(y_pred, y)
    print('epoch:', i, "loss:",loss.item())
    
    losses.append(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```
